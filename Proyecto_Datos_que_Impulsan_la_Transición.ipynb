####<font color='GREENN'>Se seleccionan las columnas que contienen información relevante para el estudio, reduciendo así el tamaño del dataframe</font>

```python
data_estudio = data[["DEPARTAMENTO", "CLASE", "COMBUSTIBLE", "AÑO_REGISTRO"]]
data_estudio
```

####<font color='GREENN'>Del conjunto de datos (data_estudio) se toman los datos de los vehículos de Risaralda</font>

---



```python
grupos = data_estudio.groupby('DEPARTAMENTO')
datos_risaralda = grupos.get_group('Risaralda')
datos_risaralda
```

####<font color='GREENN'>Se utilizó el método info() para datos_risaralda, para visualizar un resumen conciso del DataFrame, mostrando información sobre las columnas, tipos de datos, valores no nulos y uso de memoria</font>

```python
datos_risaralda.info()
```

####<font color='GREENN'>Del conjunto de datos (data_estudio) se toman los datos de los vehículos de Caldas</font>

```python
grupos = data_estudio.groupby('DEPARTAMENTO')
datos_caldas = grupos.get_group('Caldas')
datos_caldas
```

####<font color='GREENN'>Se utilizó el método info() para datos_caldas, para visualizar un resumen conciso del DataFrame, mostrando información sobre las columnas, tipos de datos, valores no nulos y uso de memoria</font>

```python
datos_caldas.info()
```

####<font color='GREENN'>Del conjunto de datos (data_estudio) se toman los datos de los vehículos de Quindio</font>

```python
grupos = data_estudio.groupby('DEPARTAMENTO')
datos_quindio = grupos.get_group('Quindio')
datos_quindio
```

####<font color='GREENN'>Se utilizó el método info() para datos_quindio, para visualizar un resumen conciso del DataFrame, mostrando información sobre las columnas, tipos de datos, valores no nulos y uso de memoria</font>

```python
datos_quindio.info()
```

####<font color='GREENN'>A continuacion para unir o concatenar las tres departamentos que son los datos sobre los cuales trabajaremos se emplea el metodo pd.concat()</font>

```python
data_eje = pd.concat([datos_quindio, datos_caldas, datos_risaralda])
data_eje
```

####<font color='GREENN'>Se utilizó el método info() para data_eje, para visualizar un resumen conciso del DataFrame, mostrando información sobre las columnas, tipos de datos, valores no nulos y uso de memoria</font>

```python
data_eje.info()
```

####<font color='GREENN'>Se saco el promedio, la moda y la madiana en año registro


```python
promedio_año_registro = data_eje['AÑO_REGISTRO'].mean()
mediana_año_registro = data_eje['AÑO_REGISTRO'].median()
moda_año_registro = data_eje['AÑO_REGISTRO'].mode()

print(f"Promedio  AÑO_REGISTRO: {promedio_año_registro}")
print(f"Mediana of AÑO_REGISTRO: {mediana_año_registro}")
print(f"Moda of AÑO_REGISTRO: {moda_año_registro.tolist()}")
```

####<font color='GREENN'>Se contaron los vehículos registrados por año

```python
contar_años = data_eje['AÑO_REGISTRO'].value_counts().sort_index()
contar_años
```

####<font color='GREENN'>Número de Vehículos Registrados por Año en Eje Cafetero</font>

```python
contar_años = data_eje['AÑO_REGISTRO'].value_counts().sort_index()

# Creación del gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(contar_años.index, contar_años.values)
barras = plt.bar(contar_años.index, contar_años.values)

plt.xlabel('Año de Registro')
plt.ylabel('Número de Vehículos')
plt.title('Número de Vehículos Registrados por Año en Eje Cafetero')
plt.xticks(contar_años.index)
plt.grid(axis='y', linestyle='--',color = 'green') # muestra o oculta líneas de cuadrícula en un gráfico
plt.bar_label(barras, padding=3)

plt.show()
```

####<font color='GREENN'>Gráfico de líneas que muestra la tendencia del número de vehículos registrados por año en el Eje Cafetero

```python
contar_años = data_eje['AÑO_REGISTRO'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.plot(contar_años.index, contar_años.values, marker='o')
plt.xlabel('Año de Registro')
plt.ylabel('Número de Vehículos')
plt.title('Tendencia del Número de Vehículos Registrados por Año en Eje Cafetero')
plt.xticks(contar_años.index)
plt.grid(True, linestyle='--', color = 'green') # muestra o oculta líneas de cuadrícula en un gráfico
for x, y in zip(contar_años.index, contar_años.values): #Con el for en "zip" se recorre el año y el valor.
    plt.text(x, y + 20, str(y), ha='center', fontsize=9, color="black") #Con el +20 se colocan las etiquetas desplazadas.

plt.show()
```

####<font color='GREENN'>Gráfico de barras que compara el número de vehículos registrados en Risaralda, Caldas y Quindío

```python
dpto_contar = data_eje['DEPARTAMENTO'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(dpto_contar.index, dpto_contar.values)
plt.xlabel('Departamento')
plt.ylabel('Número de Vehículos')
plt.title('Número de Vehículos Registrados por Departamento en Eje Cafetero')
plt.grid(axis='y', linestyle='--', color= 'green')
plt.show()
```

####<font color='GREENN'>Gráfico de barras que muestra la distribución de los tipos de combustible en el Eje Cafetero

```python
import matplotlib.pyplot as plt

# Count the occurrences of each fuel type in the combined dataframe
fuel_counts = data_eje['COMBUSTIBLE'].value_counts()

# Create the bar plot for fuel distribution
plt.figure(figsize=(10, 6))
plt.bar(fuel_counts.index, fuel_counts.values)
plt.xlabel('Tipo de Combustible')
plt.ylabel('Número de Vehículos')
plt.title('Distribución de Tipos de Combustible en Eje Cafetero')
plt.grid(axis='y', linestyle='--', color= 'green')
plt.show()
```

```python
clase_contar = data_eje['CLASE'].value_counts()

plt.figure(figsize=(12, 6))
plt.bar(clase_contar.index, clase_contar.values)
plt.xlabel('Clase de Vehículo')
plt.ylabel('Número de Vehículos')
plt.title('Distribución de Clases de Vehículos en Eje Cafetero')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', color= 'green')
plt.tight_layout() #Ajustar el diseño para evitar que las etiquetas se superpongan
plt.show()
```

```python
clase_contar = data_eje['CLASE'].value_counts()

plt.figure(figsize=(10,10))
plt.pie(clase_contar.values, labels=clase_contar.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución Porcentual de Clases de Vehículos en Eje Cafetero')
plt.axis('equal') #se dibuja como un circulo
plt.show()
```
