###<font color='GREENN'>ANÁLISIS

####<font color='GREENN'>Se aplica el filtro a "data_filtrada" para que muestre solo los valores de AUTOMÓVIL, CAMIONETA, CAMPERO Y MOTOCICLETA

```python
# isin es un método que sirve para verificar si cada valor de una columna está dentro de una lista o conjunto de valores.
data_eje_filtrada = data_filtrada[data_filtrada["CLASE"].isin(["AUTOMOVIL", "CAMIONETA", "CAMPERO", "MOTOCICLETA"])]
data_eje_filtrada
```

####<font color='GREENN'>Se implemeto una tabla dinámica para mostrar la desviación estándar y la varianza de los años de registro por departamento y tipo de combustible

```python
tabla_din = pd.pivot_table(data_eje_filtrada,
                          index=['DEPARTAMENTO', 'COMBUSTIBLE'],
                          values=['AÑO_REGISTRO'],
                          aggfunc=['std','var'])

tabla_din.rename(columns={'std': 'Desviación Estándar', 'var': 'Varianza'},
                 level= 0, inplace=True)

tabla_din.rename(columns={'AÑO_REGISTRO': ''},
                 level= 1, inplace=True)

tabla_din.columns = tabla_din.columns.set_names([None, 'AÑO DE REGISTRO'])
tabla_din
```

####<font color='GREENN'> Media móvil por año

```python
# Cuenta el número de vehículos por año
cuenta_anual = data_eje_filtrada['AÑO_REGISTRO'].value_counts().sort_index()

# Calcula el promedio móvil cada 2 años
promedio_movil = cuenta_anual.rolling(window=2).mean()

print("Promedio móvil de 2 años de vehículos matriculados:")
print(promedio_movil)
```

####<font color='GREENN'>Gráfico de línea y así visualizar el promedio móvil de 2 años del número de vehículos registrado

```python
plt.figure(figsize=(10, 6))
plt.plot(promedio_movil.index, promedio_movil.values, marker='o', linestyle='-')
plt.xlabel('Año de Registro')
plt.ylabel('Promedio Móvil de Vehículos Registrados (2 años)')
plt.title('Promedio Móvil de 2 años de Vehículos Registrados por Año en Eje Cafetero')
plt.xticks(promedio_movil.index)
plt.grid(True, linestyle='--', color = 'green')

for x, y in zip(promedio_movil.index[1:], promedio_movil.values[1:]):
    plt.text(x, y, f'{y:.1f}', ha='center', va='bottom')

plt.show()
```

####<font color='GREENN'>El promedio móvil mensual.

Para analizar los datos por mes, necesitaríamos extraer la información del mes y el año de la columna FECHA_REGISTRO y luego agrupar por esa nueva columna</font>

```python
# Cuenta el número de vehículos por año
data_eje_filtrada['FECHA_REGISTRO'] = pd.to_datetime(data_eje_filtrada['FECHA_REGISTRO'])
data_eje_filtrada['AÑO_MES'] = data_eje_filtrada['FECHA_REGISTRO'].dt.to_period('M')

cuenta_mensual = data_eje_filtrada['AÑO_MES'].value_counts().sort_index()

data_eje_filtrada['MES_REGISTRO'] = data_eje_filtrada['FECHA_REGISTRO'].dt.month #Aqui sacamos una columna dedicada a contar el mes

# Calcula el promedio móvil cada 2 meses
promedio_movil_mensual = cuenta_mensual.rolling(window=2).mean()

print("Promedio móvil de 2 meses de vehículos matriculados:")
print(promedio_movil_mensual)
```
